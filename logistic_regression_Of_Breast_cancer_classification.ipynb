## Importing the libraries

```python
import pandas as pd
```

```python
dataset = pd.read_csv('breast_cancer.csv')


```

## Importing the dataset

```python
dataset
```

```python
dataset.columns , dataset['Class'].unique()
```

```python
# dataset = pd.read_csv('breast_cancer.csv')
X = dataset.iloc[:, 1:-1].values # code number is not needed.
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=0)
```

```python
# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Logistic Regression model on the Training set

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0)
clf.fit(X_train, y_train)

print(clf)
clf,clf.coef_ , clf.penalty , clf.n_iter_, clf.intercept_
```

## Predicting the Test set results

```python
y_pred = clf.predict(X_test)
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

## Computing the accuracy with k-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = clf, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = LogisticRegression(random_state=0), X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
